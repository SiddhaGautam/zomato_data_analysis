```python
#now find all those cuisines which are most ordered or popular based on their respective country,votes and aggregate rating

sorted_cuisines=new_df.groupby(['Country','Cuisines']).agg({'Votes':'sum','Aggregate rating':'mean'}).reset_index()
```

```python
filtered_cuisines=sorted_cuisines[sorted_cuisines['Votes']*sorted_cuisines['Aggregate rating']>=2250]
```

```python
filtered_cuisines
```

```python
best_cuisines=filtered_cuisines.groupby('Country').head()
```

```python
best_cuisines
```

## so these all are top 5 foods of each country which are most popular based on their aggreagate rating and total votes

```python
## to find their average cost for two

best_cuisines_avg_cost=new_df.merge(best_cuisines[['Country','Cuisines']],on=['Country','Cuisines'],how='inner')
```

```python
avg_cost=best_cuisines_avg_cost.groupby(['Country','Cuisines','Currency'])['Average Cost for two'].mean().reset_index()
```
